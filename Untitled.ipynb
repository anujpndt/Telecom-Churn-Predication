```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # For creating plots
import matplotlib.ticker as mtick # For specifying the axes tick format
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df2 = pd.read_csv('Telco_Customer_Churn.csv')
```

```python
df2.drop('customerID',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
le = LabelEncoder()

# Apply label encoding to categorical columns
for col in df2.select_dtypes(include=['object']).columns:
    df2[col] = le.fit_transform(df2[col])

# Check transformed data
df2.head()
```

```python
X = df2.drop('Churn', axis=1)
y = df2['Churn']
```

```python
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Apply Random Undersampling
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X, y)

# Check new class distribution
print("After Undersampling:", Counter(y_resampled))

# Split into training and test sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)

# Standardize features (SVM works better with scaled data)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Initialize SVM with RBF Kernel
svm_model = SVC(kernel='rbf', probability=True, random_state=42)

# Train the model
svm_model.fit(X_train, y_train)
```

```python
# Predict on test data
y_pred = svm_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

# Print classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("SVM Confusion Matrix After Undersampling")
plt.show()
```

```python
# Apply the same scaling as training data
#X_new = X_test.iloc[[0,1,2,3,4]]  # Selecting specific rows
# Correct way to select specific rows from a NumPy array
X_new = X_test[:5]  # Selecting first 5 rows
X_new_scaled = scaler.transform(X_new)  

#X_new = X_test.sample(n=50, random_state=42)
X_new_scaled = scaler.transform(X_new)  

# Predict churn (probabilities)
y_new_prob = svm_model.predict(X_new_scaled)  

# Convert probabilities to binary values (0 or 1)
y_new_pred = (y_new_prob > 0.5).astype("int32")  

print("Predicted Churn Output:", y_new_pred.flatten())  # Convert array to readable format
print("Churn Probability:", y_new_prob.flatten())  # Probabilities for each customer
```
